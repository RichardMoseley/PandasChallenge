```python
# Dependencies and Setup
import pandas as pd

# File to Load
readFile = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purData = pd.read_csv(readFile)
```

## Player Count

* Display the total number of players


```python
#Establish a DataFrame
players = pd.DataFrame(purData)
#Find the total number of players
totalPlayers = len(players["SN"].value_counts())
#Create a DataFrame for display
totalPlayersDF = pd.DataFrame({'Players': [totalPlayers]})
#Display the data
totalPlayersDF
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Basic summary of the data
players.describe()
#Finding the number of unique items
unqItems = len(players['Item ID'].value_counts())
#Finding the average price of items and format
avgPrice = round(players['Price'].mean(), 2)
avg = "${:,.2f}".format(avgPrice)
#Finding the total number of purchases made
totalPurchases = len(players['Item ID'])
#Finding the total revenue generated from purchases and formatting
totalRev = sum(players['Price'])
revenue = "${:,.2f}".format(totalRev)
#Building a DataFrame for display purposes
summaryDF = pd.DataFrame({'Number of Unique Items': [unqItems], 'Average Price': [avg],
                          'Number of Purchases': [totalPurchases], 'Total Revenue': [revenue]})
#Displaying the data
summaryDF
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Count of Players by gender
genderGroup = players.groupby('Gender')
genderCount = genderGroup['SN'].nunique()
#Percentage of players by gender
genderPerc = genderCount/totalPlayers
#Building a dataframe for visual and formatting
genderDF = pd.DataFrame({'Total Count': genderCount, 'Percentage of Players': genderPerc})
genderDF['Percentage of Players'] = genderDF['Percentage of Players'].map("{:.2%}".format)
genderDF
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Count of Purchases by gender
purchCount = genderGroup['SN'].count()
#Percentage of purchases by gender
purchAvg = genderGroup['Price'].mean()
#Total purchase value
purchTotal = genderGroup['Price'].sum()
#Purchases per person
purchPers = purchTotal / genderGroup["SN"].nunique()
#Building a dataframe for visual and formatting
purchDF = pd.DataFrame({'Purchase Count': purchCount, 'Average Purchase Price': purchAvg, 
                        'Total Purchase Value': purchTotal, 'Avg Total Purchase per Person': purchPers})
purchDF['Average Purchase Price'] = purchDF['Average Purchase Price'].map("${:.2f}".format)
purchDF['Total Purchase Value'] = purchDF['Total Purchase Value'].map("${:.2f}".format)
purchDF['Avg Total Purchase per Person'] = purchDF['Avg Total Purchase per Person'].map("${:.2f}".format)
purchDF
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Creating bins
playersCopy = players.copy()
newDF = playersCopy.drop_duplicates(subset=['SN'])
ageBins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 100]
binNames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#Separating into bins
newDF['Sorted Age'] = pd.cut(newDF['Age'], ageBins, labels=binNames)
#Finding total players
ageTotals = newDF['Sorted Age'].value_counts()
#Finding percentages
agePercs = ageTotals / totalPlayers
#Creating dataframe and formatting
ageDF = pd.DataFrame({'Total Count': ageTotals, 'Percentage of Players': agePercs})
ageDF['Percentage of Players'] = ageDF['Percentage of Players'].map("{:.2%}".format)
ageDF.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Separating into bins
players['Sorted Purch'] = pd.cut(players['Age'], ageBins, labels=binNames)
#Grouping my bins
playersGroup = players.groupby('Sorted Purch')
#Finding the total purchases
purchCount = players['Sorted Purch'].value_counts()
#Finding the purchase total
purchTotal = playersGroup['Price'].sum()
#Finding the purchase average
purchAvg = playersGroup['Price'].mean()
#Finding the average per person
purchPer = purchTotal / purchCount
#Building the DataFrame and formatting
purchDF = pd.DataFrame({'Purchase Count': purchCount, 'Average Purchase Price': purchAvg,
                        'Total Purchase Value': purchTotal, 'Avg Total Purchase per Person': purchPer})
purchDF['Average Purchase Price'] = purchDF['Average Purchase Price'].map("${:.2f}".format)
purchDF['Total Purchase Value'] = purchDF['Total Purchase Value'].map("${:.2f}".format)
purchDF['Avg Total Purchase per Person'] = purchDF['Avg Total Purchase per Person'].map("${:.2f}".format)
purchDF
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Grouping by SN
snGroup = players.groupby('SN')
#Finding the total player purchases
snPurchCount = players['SN'].value_counts()
#Finding the average player purchase
snAvgPurch = snGroup['Price'].mean()
#Finding the sum of all player purchases
snTotalPurch = snGroup['Price'].sum()
#Building a DataFrame and formatting
snDF = pd.DataFrame({'Purchase Count': snPurchCount,
                    'Average Purchase Price': snAvgPurch,
                    'Total Purchase Value': snTotalPurch})
sortedDF = snDF.sort_values(by='Total Purchase Value', ascending=False)
sortedDF['Average Purchase Price'] = sortedDF['Average Purchase Price'].map("${:.2f}".format)
sortedDF['Total Purchase Value'] = sortedDF['Total Purchase Value'].map("${:.2f}".format)
sortedDF.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
itemGroup = players.groupby(['Item ID', 'Item Name'])
itemCount = itemGroup['Purchase ID'].count()
itemTotal = itemGroup['Price'].sum()
itemPrice = itemTotal / itemCount
itemDF = pd.DataFrame({'Purchase Count': itemCount,
                      'Item Price': itemPrice,
                      'Total Purchase Value': itemTotal})
itemsortDF = itemDF.sort_values(by='Purchase Count', ascending=False)
itemsortDF['Item Price'] = itemsortDF['Item Price'].map("${:.2f}".format)
itemsortDF['Total Purchase Value'] = itemsortDF['Total Purchase Value'].map("${:.2f}".format)
itemsortDF.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
newsortDF = itemDF.sort_values(by='Total Purchase Value', ascending=False)
newsortDF['Item Price'] = newsortDF['Item Price'].map("${:.2f}".format)
newsortDF['Total Purchase Value'] = newsortDF['Total Purchase Value'].map("${:.2f}".format)
newsortDF.head()
```
